```python
dias= list(range(len(estaciones)))
```

```python
# Crear el dataframe
df = pd.DataFrame({
    'Estacion': estaciones,
    'Personas': personas_que_entraran,
    'Lloviendo': lluvia_num, 
    'Dia':dias
})
```

```python
df_onehot=pd.get_dummies(df, columns=['Estacion'])
```

```python
# Separar X (variables independientes) e y (variable dependiente)
X = df_onehot.drop('Personas', axis=1)
y = df_onehot['Personas']
```

```python
modelo = LinearRegression()
modelo.fit(X, y)
```

```python
#creamos la grafica
fig = plt.figure(figsize=(20, 8))#tam de grafica
ax = fig.add_subplot(111, projection='3d') #tipo de grafico 3D
#graficar los datos 
xx =list(range(len(df['Estacion'])))
ax.scatter(xx, df['Lloviendo'], y, color='blue', label='personas')


ax.set_xticks(xx[::10])  # una cada 10 estaciones
ax.set_xticklabels([estaciones[i] for i in xx[::10]], rotation=90)


x1_surface, x2_surface = np.meshgrid(
    np.linspace(0, len(estaciones) - 1, 10),  # estaciones como √≠ndices num√©ricos
    np.linspace(0, 1, 10)                      # lluvia (0 o 1)
)
Z = modelo.intercept_ + modelo.coef_[0]*x1_surface + modelo.coef_[1]*x2_surface
ax.plot_surface(x1_surface, x2_surface, Z, alpha=0.5, color='green')

# Etiquetas
ax.set_xlabel('Estaci√≥n ')
ax.set_ylabel('¬øLluvia?')
ax.set_zlabel('Personas')
#titulo
plt.title('Regresi√≥n lineal m√∫ltiple')
plt.legend()
plt.show()
```
